```python
def OLS_fit(X_T,y):
    """Ordinary Least Square fit of Linear Regression
    Args:
        X: list of X array
        y: list of y array
    Returns:
        coef: coefficient of linear regression
    Raises:
        AssertionError: if the length of X and y doesn't match
    """
    assert len(X_T)==len(y)
    X_T=X_T.to_numpy()
    y=y.to_numpy()
    X_T.resize(len(X_T),1)
    y.resize(len(y),1)
    
    X= np.transpose(X_T)
    print(X.shape)
    coef=np.matmul(X_T,X)
    print(coef.shape)
    Pseudo=np.linalg.pinv(coef)
    print(Pseudo.shape)
    
    #coef= np.dot(np.dot(inv(np.dot(X_T,X)),X_T),y)
    return coef
```
